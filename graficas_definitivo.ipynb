```python
df_demanda1=pd.DataFrame()
df_demanda1['Suma']=df_demanda['Suma']
df_demanda1['Date']=df_demanda['Date']
df_demanda1['promedio'] = df_demanda1.groupby(df_demanda1['Date'].dt.to_period("M")).transform('mean').mean(axis=1)
```

```python
df_demanda1
```

### Aportes

```python
objetoAPI = pydataxm.ReadDB()
df_aporte = objetoAPI.request_data(
                        "AporEner",
                        "Sistema",
                        dt.date(2023,9,7),
                        dt.date(2023,12,6))
df_aporte
```

Conversion de dato de fecha y el promedio de aporte por mes (aparte de la trasformación a GWh)

```python
df_aporte['Date'] = pd.to_datetime(df_aporte['Date'])
promAporte = df_aporte.groupby(df_aporte['Date'].dt.to_period("M")).transform('mean').mean(axis=1)
promAporte= promAporte/1000000
```

Media historica del aporte

```python
objetoAPI = pydataxm.ReadDB()
df_aporteh = objetoAPI.request_data(
                        "AporEnerMediHist",
                        "Sistema",
                        dt.date(2023,9,7),
                        dt.date(2023,12,6))
df_aporteh
```

### Creacion de la grafica

```python
#plt.style.use('bmh')
fecha = pd.to_datetime(df_aporte['Date'], format="%Y-%m-%d %H:%M:%S")


x= fecha.dt.strftime("%b-%d")
aporte = df_aporte['Value']/1000000
aporteM = df_aporteh['Value']/1000000
Demanda = df_demanda1['Suma']/1000000
DemandaM = df_demanda1['promedio']/1000000

porcProm =  float(round(((aporteM.iloc[-1] - promAporte.iloc[-1])/aporteM.iloc[-1])*100))
porcAjus =  float(round(((aporteM.iloc[-1] - aporte.iloc[-1])/aporteM.iloc[-1])*100))


fig, ax = plt.subplots(1, 1, figsize=(20,8))

ax.plot(x,aporte,linestyle='--',color='lightblue')
ax.plot(x,aporteM,label='Aportes Medios',color='navy')
ax.plot(x,Demanda,label='Demanda Nacional',color='yellowgreen')
ax.plot(x,DemandaM,label='Demanda Media',color='darkorange')


ax.set_title('Aportes vs Demandas', fontsize=26,fontfamily='serif',pad=50,color='dimgray',loc='center')
ax.fill_between(x,aporte, color='lightblue', alpha=0.4,label='Aportes Energía')
plt.xticks(range(0, len(x), 2), x[::2], rotation=45,color='dimgray',fontweight='bold')

plt.text(x.iloc[0], 75,f'Demanda Dia: {float(round(Demanda.iloc[-1]))} GWh/dia {x.iloc[-1]}',color='g',fontfamily='serif',fontsize=13,fontweight='bold')
plt.text(x.iloc[0], 25,f'Demanda Mes: {float(round(DemandaM.iloc[-1]))} GWh/dia promedio {x.iloc[-1]}',color='darkorange',fontfamily='serif',fontsize=13,fontweight='bold')
plt.text(x.iloc[55], 290,f'Aportes del día: {float(round(aporte.iloc[-1]))} GWh/dia {x.iloc[-1]}',color='navy',fontfamily='serif',fontsize=13,fontweight='bold')
plt.text(x.iloc[52], 240,f'{porcAjus} % respecto a la media historica [{float(round(aporteM.iloc[-1]))} GWh/día]',color='b',fontfamily='serif',fontsize=13,fontweight='bold')
plt.text(x.iloc[51], 75,f'Aportes Mes: {float(round(promAporte.iloc[-1]))} GWh/dia promedio {x.iloc[-1]}',color='b',fontfamily='serif',fontsize=13,fontweight='bold')
plt.text(x.iloc[45], 25,f'{porcProm} % respecto a la media historica [{float(round(aporteM.iloc[-1]))} GWh/día]',color='b',fontfamily='serif',fontsize=13,fontweight='bold')


ax.set_ylabel('GWh/dia',fontsize=16,fontfamily='serif',color='gray',labelpad=20,)
ax.legend(loc='upper left', fontsize=12, frameon=False,ncol=5)

img = Image.open('Logo Rectangular SFE.jpg')

size = (240, 110)
img = img.resize(size)

fig.figimage(img, 100, 1150, alpha=0.4)

plt.show()
```

# **Termicas**

### Producción Termica

```python
objetoAPI = pydataxm.ReadDB()
df = objetoAPI.request_data(
                        "Gene",
                        "Recurso",
                        dt.date(2023,11,7),
                        dt.date(2023,12,6))
df
```

### Listado de Plantas

```python
df_sistema = objetoAPI.request_data(
                                'ListadoRecursos',
                                'Sistema',
                                dt.date(2023, 11, 7),
                                dt.date(2023, 12, 6))
df_sistema
```

Se eliminan las variables innecesarias

```python
df_sistema = df_sistema[['Values_Code', 'Values_Type', 'Values_EnerSource']]
```

Se juntan las dos tablas

```python
df = pd.merge(df,df_sistema,left_on=['Values_code'],right_on=['Values_Code'],how='left')
df
```

Se filtran por termicas

```python
df_termica = df.query('Values_Type == "TERMICA"')
df_termica
```

```python
set(df_termica['Values_EnerSource'].values)
```

Se filtran por tipo de recurso

```python
gas=['GAS','BIOGAS']
liquido = ['COMBUSTOLEO','JET-A1','ACPM','BAGAZO',]
gases = df.query("Values_EnerSource in @gas")
liquidos = df.query("Values_EnerSource in @liquido")
carbon = df.query("Values_EnerSource == 'CARBON'")
```

Se eliminan variables innecesarias y se eliminan nulos

```python
liquidos1=liquidos.drop(['Id','Values_code','Date','Values_Type','Values_EnerSource','Values_Code'],axis=1)
gases1=gases.drop(['Id','Values_code','Date','Values_Type','Values_EnerSource','Values_Code'],axis=1)
carbon1=carbon.drop(['Id','Values_code','Date','Values_Type','Values_EnerSource','Values_Code'],axis=1)
liquidos1 = liquidos1.fillna(0)
gases1 = gases1.fillna(0)
carbon1 = carbon1.fillna(0)
```

```python
liquidos1['Suma'] = liquidos1.sum(axis=1)
gases1['Suma'] = gases1.sum(axis=1)
carbon1['Suma'] = carbon1.sum(axis=1)
liquidos1['fecha'] = liquidos['Date']
gases1['fecha'] = gases['Date']
carbon1['fecha'] = carbon['Date']
```

Se suman por dias

```python
carbon1=carbon1[['fecha', 'Suma']]
liquidos1=liquidos1[['fecha', 'Suma']]
gases1=gases1[['fecha', 'Suma']]
carbon2 = carbon1.groupby('fecha')['Suma'].sum().reset_index()
liquidos2 = liquidos1.groupby('fecha')['Suma'].sum().reset_index()
gases2 = gases1.groupby('fecha')['Suma'].sum().reset_index()
carbon2['Suma']=carbon2['Suma']/1000000
liquidos2['Suma']=liquidos2['Suma']/1000000
gases2['Suma']=gases2['Suma']/1000000
```

```python
gases2
```

Se crea la grafica

```python

fecha = pd.to_datetime(gases2['fecha'], format="%Y-%m-%d %H:%M:%S")

x= fecha.dt.strftime("%b-%d")



fig, ax = plt.subplots(1, 1, figsize=(25,10))
ax.bar(x, carbon2['Suma'], label='Gen. Carbon',color='maroon',width=0.6)
ax.bar(x, gases2['Suma'], bottom=carbon2['Suma'], label='Gen. Gas',color='darkorange',width=0.6)
ax.bar(x, liquidos2['Suma'], bottom=[i + j for i, j in zip(carbon2['Suma'], gases2['Suma'])], label='Gen. Liquido',color='r',width=0.6)
ax.set_xticks(x)
ax.set_xticklabels(x, rotation=45, ha='right',fontsize=16,color='gray')
ax.set_title('Despacho vs Generación Termica', fontsize=36,fontfamily='serif',pad=50,color='gray',loc='center',horizontalalignment='center')

for i in range(len(x)):
  ax.text(x.iloc[i],(carbon2['Suma'][i])/2,round(carbon2['Suma'][i],2),rotation='vertical',color='white',fontsize=16,fontweight='bold',horizontalalignment='center',verticalalignment='center')
  ax.text(x.iloc[i],(carbon2['Suma'][i]+(gases2['Suma'][i]/2)),round(gases2['Suma'][i],2),rotation='vertical',color='white',fontsize=16,fontweight='bold',horizontalalignment='center',verticalalignment='center')
  ax.text(x.iloc[i],(gases2['Suma'][i]+carbon2['Suma'][i]+(liquidos2['Suma'][i]/2)),round(liquidos2['Suma'][i],2),rotation='vertical',color='white',fontsize=14,fontweight='bold',horizontalalignment='center',verticalalignment='center')

#ax.set_xlabel('Fecha',labelpad=20,fontsize=16,fontfamily='serif',color='gray')
ax.set_ylabel('GWh/Dia',fontsize=20,fontfamily='serif',color='dimgray')
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3,frameon=False,fontsize=18)

img = Image.open('Logo Rectangular SFE.jpg')

size = (240, 110)
img = img.resize(size)

fig.figimage(img, 100,950, alpha=0.4)

plt.show()
```
